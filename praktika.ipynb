```python
class gyvunas:
    def __init__(self, vardas, amzius, svoris):
        self.vardas = vardas
        self.amzius = amzius
        self.svoris = svoris

    def __str__(self):
            return f'{self.vardas} {self.amzius}: {self.svoris}'
        
def padidinti_svori(self, svoris):
    svoris = self.svoris + svoris
    self.svoris = svoris
    return f"bendras svoris: {svoris}"
        
def numete_svorio(self, svoris):
    svoris = self.svoris - svoris
    self.svoris = svoris
    return f"bendras svoris: {svoris}"

def gimtadienis(self):
    self.amzius = self.amzius + 1
    return f"Su gimtadieniu {self.vardas}.Tau suejo {self.amzius} metai."
    
    
maikis = gyvunas("Maikis", 1, 1.3)    
maikis.vardas = "Maikis"
maikis.amzius = 1
maikis.svoris = 1.3

print(padidinti_svori(maikis, 2))
print(numete_svorio(maikis, 2))
print(gimtadienis(maikis))
```

```python
class BankoSaskaita:
    def __init__(self, saskaitos_nr, savininkas, balansas = 500, pin_kodas = "1233", bandymu_skaicius = 3):
        self.saskaitos_nr = saskaitos_nr
        self.savininkas = savininkas
        self.__balansas = balansas
        self.__pin_kodas = pin_kodas
        self.__bandymu_skaicius = bandymu_skaicius

    def saskaitos_likutis(self):
        return self.__balansas

    def pin_kodo_patikrinimas(self, pin_kodas):
        if self.__bandymu_skaicius > 0:
            if self.__pin_kodas == pin_kodas:
                self.__bandymu_skaicius = 3
                return True
            else:
                print("neteisingas PIN kodas")
                self.__bandymu_skaicius -= 1
        else:
            print("kortele uzblokuota!")

    def __keisti_balansa(self, pinigai):
        self.__balansas += pinigai
        return pinigai
    
    def nusiimti_pingigu(self, pinigai, pin_kodas):
        if self.pin_kodo_patikrinimas(pin_kodas):
            return self.__keisti_balansa(-pinigai)
    
    def prideti_pingigu(self, pinigai, pin_kodas):
        if self.pin_kodo_patikrinimas(pin_kodas):
            return self.__keisti_balansa(pinigai)
            
klientas = BankoSaskaita(1, "Martynas")
print(klientas.nusiimti_pingigu(200, "1232"))
print(klientas.nusiimti_pingigu(200, "1232"))
print(klientas.prideti_pingigu(500, "1233"))
print(klientas.prideti_pingigu(500, "12333"))
print(klientas.prideti_pingigu(500, "12333"))
print(klientas.prideti_pingigu(500, "12333"))
print(klientas.prideti_pingigu(500, "12333"))
```

```python
class Knygos:
    def __init__(self, pavadinimas, autorius, puslapiu_skaicius = 500, bukle = "prasta"):
        self.__pavadinimas = pavadinimas
        self.__autorius = autorius
        self.__puslapiu_skaicius = puslapiu_skaicius
        self.__bukle = bukle
    
    def __knygos_pavadinimas(self):
        return self.__pavadinimas
    
    def __knygos_autorius(self):
        return self.__autorius
    
    def __knygos_puslapiai(self):
        return self.__puslapiu_skaicius
    
    def __knygos_bukle(self):
        return self.__bukle
    
    def pakeisti_knygos_bukle(self, bukle):
        if bukle in ("patenkinama", "prasta", "atnaujinta", "sugadinta"):
            self.__knygos_bukle = bukle
            return bukle
        else:
            print("tokia bukle nepriimtina")

    def __naujas_puslapiu_kiekis(self, puslapiai):
        self.__knygos_puslapiai = puslapiai
        return self.__knygos_puslapiai

    def isplesti_knygos_puslapius(self, puslapiai):
        if abs(puslapiai) <= self.__knygos_puslapiai():
            return self.__naujas_puslapiu_kiekis(self.__knygos_puslapiai() - abs(puslapiai))
        else:
            print("per didelis kiekis")
    
gyvaciu_karas = Knygos("gyvaciu karas", "Alma Litera")

print(gyvaciu_karas.pakeisti_knygos_bukle("atnaujinta"))
print(gyvaciu_karas.isplesti_knygos_puslapius(200))
```

```python
class Baseinas:
    def __init__(self, vandens_talpa, kiek_liko_vandens):
        self.vandens_talpa = vandens_talpa
        self.kiek_liko_vandens = kiek_liko_vandens
    
    def __kiek_yra_vandens(self):
        kiek_yra_vandens = self.kiek_liko_vandens
        return kiek_yra_vandens
    
    def pripilti_baseina(self, kiek_pilti=10):
        if kiek_pilti + self.__kiek_yra_vandens() > self.vandens_talpa:
            print("tiek ipilti, kad tilptu i baseina neimanoma")
        else:
            self.kiek_liko_vandens = kiek_pilti + self.__kiek_yra_vandens()
            print(f"ipylem: {kiek_pilti}, is viso yra vandens: {self.kiek_liko_vandens}")

    def isleisti_baseina(self, kiek_isleisti=10):
        if self.__kiek_yra_vandens() - kiek_isleisti < 0:
            print("neimanoma ispilti tiek vandens")
        else:
            self.kiek_liko_vandens = self.__kiek_yra_vandens() - kiek_isleisti
            print(f"baseine liko: {self.kiek_liko_vandens}")

mazas_baseinas = Baseinas(2000, 500)
print(mazas_baseinas.pripilti_baseina())
print(mazas_baseinas.pripilti_baseina(500))
print(mazas_baseinas.pripilti_baseina(990))
print(mazas_baseinas.pripilti_baseina(1))
print(mazas_baseinas.isleisti_baseina(500))
print(mazas_baseinas.isleisti_baseina(1500))
print(mazas_baseinas.isleisti_baseina(1))
```

```python
class Darbuotojas():
    def __init__(self, vardas, pavarde, atlyginimas):
        self.vardas = vardas
        self.pavarde = pavarde
        self.atlyginimas = atlyginimas
    
    def informajica(self):
        print(f"Vardas: {self.vardas}\nPavarde: {self.pavarde}\nAtlyginimas: {self.atlyginimas}")
    
mazvydas = Darbuotojas("Mazvydas", "Pavardenis", 1000)
print(mazvydas.informajica())
```

```python
class Administratorius(Darbuotojas):

    def informajica(self):
        return super().informajica()

antanas = Administratorius("Antanas", "Povandeninis", 1500)
print(antanas.informajica())
```

```python
class Vadovas(Darbuotojas):
    premija = 500

    def informajica(self):
        return print(f"Premija: {self.premija}\n{super().informajica()}")
    
martyna = Vadovas("Martyna", "Paselusi", 2000)
print(martyna.informajica())
```
